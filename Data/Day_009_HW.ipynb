```python
import numpy as np

np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 隨機生成 1000 個介於 0~50 的數
x = np.random.randint(0, 50, 1000)

# 正相關，增加一些雜訊
y = x + np.random.normal(0, 10, 1000)

# 可以用 numpy 裡寫好的 function 來計算相關係數
np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```
